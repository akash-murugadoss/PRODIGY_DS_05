```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = "/content/Road Accident Data.csv"
data = pd.read_csv(file_path)

# Explore the dataset
print(data.head())
print(data.info())

# Analyze patterns related to road conditions, weather, and time of day
# Visualize accident hotspots
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=data, hue='Accident_Severity', palette='coolwarm', legend='full', alpha=0.5)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Visualize contributing factors
plt.figure(figsize=(10, 6))
sns.countplot(y='Road_Surface_Conditions', data=data, order=data['Road_Surface_Conditions'].value_counts().index, palette='viridis')
plt.title('Road Surface Conditions')
plt.xlabel('Count')
plt.ylabel('Road Surface Condition')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='Weather_Conditions', data=data, order=data['Weather_Conditions'].value_counts().index, palette='viridis')
plt.title('Weather Conditions')
plt.xlabel('Count')
plt.ylabel('Weather Condition')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='Time', data=data, order=data['Time'].value_counts().index[:10], palette='viridis')
plt.title('Top 10 Time of Day for Accidents')
plt.xlabel('Count')
plt.ylabel('Time of Day')
plt.show()
```
